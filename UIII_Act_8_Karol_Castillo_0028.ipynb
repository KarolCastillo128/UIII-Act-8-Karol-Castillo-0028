## Proyecto de simulación de inteligencia artificial con if, else, elif  while tipo Menu con ESCULTOR Y CINCEL

```python
# =========================================
# SIMULADOR DE IA: ESCULTOR Y CINCEL
# Ejemplo simple para preparatoria
# Usa: if, elif, else y while
# NO usa funciones, clases, for ni try
# =========================================

print("===================================")
print("   IA DEL ESCULTOR Y EL CINCEL")
print("===================================")
print("La inteligencia artificial intentará")
print("adivinar la mejor escultura para ti.")
print()

continuar = "si"

while continuar == "si":

    print("\nMENU DE MATERIAL")
    print("1. Piedra")
    print("2. Madera")
    print("3. Hielo")

    material = input("Elige un material: ")

    print("\nMENU DE HERRAMIENTA")
    print("1. Cincel fino")
    print("2. Cincel pesado")
    print("3. Cincel artístico")

    herramienta = input("Elige un cincel: ")

    print("\nLa IA está pensando...")
    print("-----------------------------------")

    # DECISIONES DE LA IA
    if material == "1" and herramienta == "1":
        print("La IA recomienda crear una estatua pequeña.")
        print("Precisión perfecta para detalles.")

    elif material == "1" and herramienta == "2":
        print("La IA recomienda crear un monumento gigante.")
        print("El cincel pesado rompe piedra fácilmente.")

    elif material == "1" and herramienta == "3":
        print("La IA recomienda una escultura moderna.")
        print("Arte abstracto detectado.")

    elif material == "2" and herramienta == "1":
        print("La IA recomienda tallar un animal.")
        print("La madera permite detalles suaves.")

    elif material == "2" and herramienta == "2":
        print("La IA recomienda fabricar un tótem.")
        print("Gran fuerza detectada.")

    elif material == "2" and herramienta == "3":
        print("La IA recomienda una máscara artística.")
        print("Creatividad elevada.")

    elif material == "3" and herramienta == "1":
        print("La IA recomienda un copo de nieve gigante.")
        print("Trabajo delicado sobre hielo.")

    elif material == "3" and herramienta == "2":
        print("La IA recomienda un castillo de hielo.")
        print("Construcción rápida detectada.")

    elif material == "3" and herramienta == "3":
        print("La IA recomienda una escultura futurista.")
        print("Modo creativo activado.")

    else:
        print("La IA no entiende la selección.")
        print("Debes elegir opciones válidas del menú.")

    print("-----------------------------------")

    continuar = input("\n¿Deseas probar otra escultura? (si/no): ")

print("\nGracias por usar la IA del Escultor.")
print("Programa finalizado.")
```
